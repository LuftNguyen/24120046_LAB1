```python
import os
import threading
import time

def start_pinggy():
    os.system("printf '\n' | ssh -p 443 -o StrictHostKeyChecking=no -R0:localhost:8000 a.pinggy.io > log_pinggy.txt 2>&1")

# Khởi chạy luồng (Thread)
threading.Thread(target=start_pinggy, daemon=True).start()

# Đợi 5 giây
print("Đang khởi tạo đường hầm Pinggy... Vui lòng đợi 5 giây...")
time.sleep(5)

# Đọc file log để lấy link
if os.path.exists("log_pinggy.txt"):
    with open("log_pinggy.txt", "r") as f:
        content = f.read()
        print("--- THÔNG TIN PINGGY ---")
        print(content)
else:
    print("Lỗi: Không tìm thấy file log_pinggy.txt")
```
